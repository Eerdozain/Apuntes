Primero vamos a realizar el paseo aleatorio de 1 dimensión del supuesto del fondo de la huerta.Se conoce que la probabilidad de que llueva es del 30% y de que haga sol 70%. La huerta tiene un fondo de 550.000€, si llueve tiene pérdidas del 10% y si hace sol beneficios del 20%. Ahora estudiaremos os beneficios en caso de que haga sol en un período de 50 días.

```python
import random
import matplotlib.pyplot as plt
random.seed()
fondo = 550000     # indicamos el valor inicial del fondo de la huerta
n = 50         # establecemos el número de días del estudio
factor = 1.2    # establecemos el factor. Analizamos las ganancias por lluvia, como el incremento es del 20% el factor será 1.2
p = .7           # probabilidad de subida. Si es 0.7 la probabilidad de que haga sol es del 70%
serie = [fondo] # inicializamos la serie del fondo, comenzamos con la primera cantidad del fondo
for t in range(1,n):
    if random.random() < p:
        fondo *= factor
    else:
        fondo /= factor
    serie.append(fondo)
plt.plot(range(n),serie)
plt.show()
```

Ahora pasamos al paseo aleatorio de dos dimensiones

```python
# La posibilidad de la lluvia y con ella los beneficios aportados se mueve en un paseo aleatorio en dos dimensiones
 # Se parte del punto (0,0) del plano cartesiano  
# Cada paso que se da es de longitud r  
# El ángulo con el que se da cada paso es aleatorio  
import random  
import math  
import matplotlib.pyplot as plt  
random.seed()  
n=50 # número de días  
r=1 # longitud del paso  
x=y=0 # punto de partida (x,y)=(0,0)  
listax=[x]  
listay=[y]  
for i in range(n):  
  angulo=random.random()*math.pi*2 # en radianes  
  x+=r*math.cos(angulo)  
  y+=r*math.sin(angulo)  
  listax.append(x)  
  listay.append(y)  
plt.plot(listax,listay)  
plt.show()
```
